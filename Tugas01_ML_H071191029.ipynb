```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import transactionencoder
```

```python
dataset = [('Kucing', 'Bebek', 'Ayam', 'Buaya', 'Singa'),
           ('Panda', 'Bebek', 'Ayam', 'Buaya', 'Singa'),
           ('Kucing', 'Burung', 'Ayam', 'Buaya'),
           ('Kucing', 'Domba', 'Macan', 'Ayam', 'Singa'),
           ('Macan', 'Bebek', 'Bebek', 'Ayam', 'Anjing', 'Buaya')]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder 
```

```python
dataset
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset) 
```

```python
dataset = pd.DataFrame(te_ary, columns=te.columns_)
dataset.head()
```

```python
frequent_itemset = apriori(dataset, min_support=0.6, use_colnames=True)
frequent_itemset
```

```python
result = association_rules(frequent_itemset, metric="confidence", min_threshold=0.5)
result
```

```python
result_simplify = result[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
result_simplify
```
